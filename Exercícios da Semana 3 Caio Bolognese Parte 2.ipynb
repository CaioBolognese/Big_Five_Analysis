```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# começando com o básico lendo o dataset escolhido e expondo o cabeçalho para verificar sua distribuição certinho

dataset = pd.read_csv('Dataset da Semana 3\BIG FIVE 1995-2019.csv')
dataset.head()
```

```python
# o dataset possui uma coluna denominada Country que indica o país onde foi realizada a partida e por consequência em qual liga os times pertencem
# chamando o .unique() verificamos quais países são esses. 
lista_de_ligas = dataset['Country'].unique()
lista_de_ligas
```

```python
# aqui usamos um loop for para dividir as ligas cada uma em um arquivo diferente, dividindo assim permite que a analise fique mais centrada para cada uma delas
# e também poupa trabalho se o objetivo for estudar cada uma de maneira separada
for liga in lista_de_ligas:
    caminho_do_arquivo = f"./Dataset da Semana 3/liga_{liga}.csv"
    dataset[dataset['Country'] == liga].to_csv(caminho_do_arquivo)
```

```python
# lendo o novo arquivo criado a partir da iteração anterior

premier_league = pd.read_csv('./Dataset da Semana 3/liga_ENG.csv')
premier_league
```

```python
# fica aqui já as todas as outras ligas caso necessite a utilização delas posteriormente

la_liga = pd.read_csv('./Dataset da Semana 3/liga_ESP.csv')
bundesliga = pd.read_csv('./Dataset da Semana 3/liga_GER.csv')
serie_a = pd.read_csv('./Dataset da Semana 3/liga_IT.csv')
ligue_1 = pd.read_csv('./Dataset da Semana 3/liga_FR.csv')
```

```python
# renomeando as colunas para deixar a compreensão mais facilitada e testar o . rename para saber como funciona, importante destacar o parâmetro inplace que vai ser muito utilizado

premier_league.rename(columns={'Round': 'Rodada', 'Date': 'Dia do Jogo', 'Team 1': 'Time 1', 'Team 2': 'Time 2', 'FT': 'ST', 'HT': 'PT', 'Year': 'Ano', 'Country': 'Localidade'}, inplace = True)
premier_league.head()
```

```python
# retirando um coluna desnecessária com o .drop
premier_league.drop(columns='Unnamed: 0', inplace= True)
```

```python
# senti a necessidade de separar novamente dessa vez por temporada o dataset, sinto que facilita muito a menor quantidade de dados a serem trabalhados nesse caso visto que meu objetivo
# é remontar a tabela final de apenas uma temporada

temporadas = premier_league['Ano'].unique()
temporadas
```

```python
# aplicando o mesmo conceito do loop desta vez para cada temporada

for ano in temporadas:
    caminho_do_arquivo = f"./Dataset da Semana 3/temporada_{ano}.csv"
    premier_league[premier_league['Ano'] == ano].to_csv(caminho_do_arquivo)
```

```python
# lendo novamente o dataset

temporada_2018 = pd.read_csv('./Dataset da Semana 3/temporada_2018.csv')
temporada_2018.head()
```

```python
# a mesma coluna volta a aparecer então removendo ela de novo

temporada_2018.drop(columns= 'Unnamed: 0', inplace=True)
```

```python
# aqui vamos criar as pontuações dentro e fora de casa que permitem posteriormente uma análise mais detalhada

pontuacao_em_casa = temporada_2018.groupby(['Ano','Localidade','Time 1']).sum().reset_index()[['Ano','Localidade','Time 1','Team 1 (pts)' ]]

pontuacao_fora_casa = temporada_2018.groupby(['Ano','Localidade','Time 2']).sum().reset_index()[['Ano','Localidade','Time 2','Team 2 (pts)' ]]
```

```python
# renomeando as colunas para ficar tudo em português 

pontuacao_em_casa.rename(columns = {'Time 1':'Time', 'Team 1 (pts)':'Pontuação'}, inplace=True)

pontuacao_fora_casa.rename(columns = {'Time 2':'Time', 'Team 2 (pts)':'Pontuação'}, inplace=True)
```

```python
# criando uma coluna para definir quando um jogo foi realizado dentro ou fora de cara 

pontuacao_fora_casa['Local'] = 'Fora de Casa'

pontuacao_em_casa['Local'] = 'Em Casa'
```

```python
# no dataframe final é disposto os times que participaram dessa temporada nesse ano juntamente com a pontuação obtida jogando dentro e fora de casa

contabiliza_pontos = pontuacao_em_casa.append(pontuacao_fora_casa)
contabiliza_pontos
```
